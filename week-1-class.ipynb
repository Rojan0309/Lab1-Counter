Here is a little sample of plotting timings for a function in Python.

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def fake_function(n):
    time.sleep(0.001*n*n)

def timeit(n):
    start = time.time()
    fake_function(n)
    end = time.time()
    return end-start

N = [0, 10, 20, 30, 40, 50]
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

## Implementation notes on compiled languages

For the other ones, if you're using a compiled language, they might run so fast that in order to time them you should slow down the addition by adding a 1 microsecond wait. I've included a C++ ``busy_sleep`` function below to do that.

```c++
#include <chrono>
#include <iostream>
#include <fstream>

using namespace std;

// this function busy sleeps for n microseconds
void busy_sleep(int n)
{
    auto start = chrono::steady_clock::now();
    while(chrono::duration_cast<chrono::microseconds>(chrono::steady_clock::now() - start).count()<n) {};
}
```

EXERCISE: Recursive

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def fib_recurse(n):
    if n<=2:
      return 1
    else:
      return fib_recurse(n-1)+fib_recurse(n-2)

def timeit(n):
    start = time.time()
    fib_recurse(n)
    end = time.time()
    return end-start

N = np.arange(31)
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

EXERCISE: Keep last two values

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def fib_last2(n):
   a = b = 1
   for i in range(3, n + 1):
       a, b = b, a + b
   return b

def timeit(n):
    start = time.time()
    fib_last2(n)
    end = time.time()
    return end-start

N = np.arange(101)
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

EXERCISE: Analytic

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
import math

def fib_analytic(n):
   z = (1 + math.sqrt(5))/2
   fib = (z**n-(1-z)**n)/math.sqrt(5)
   return fib

def timeit(n):
    start = time.time()
    fib_analytic(n)
    end = time.time()
    return end-start

N = np.arange(101)
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

EXERCISE: Matrix

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
import math

def fib_matrix(n):
   a = np.array([[1, 1],
                 [1, 0]])

   fib = np.linalg.matrix_power(a, n)
   return fib

def timeit(n):
    start = time.time()
    fib_matrix(n)
    end = time.time()
    return end-start

N = np.arange(101)
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````
